```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# load sample CSV (relative path)
df = pd.read_csv('sample-data/heart-rate-sample.csv')
df.head()
```

```python
# quick time-series plot (adjust column names if needed)
if 'time' in df.columns:
    df['time'] = pd.to_datetime(df['time'])
    plt.figure(figsize=(10,4))
    plt.plot(df['time'], df.iloc[:,1])
    plt.xlabel('time')
    plt.ylabel(df.columns[1])
    plt.title('Sample heart rate')
    plt.tight_layout()
    plt.show()
else:
    df.plot(figsize=(10,4))
    plt.show()
```
